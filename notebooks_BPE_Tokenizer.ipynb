```python
import numpy as np
import re
from collections import Counter

print("Libraries imported successfully!")
```

```python
corpus = [
    "low",
    "lower",
    "lowest",
    "new",
    "newer",
    "wide",
    "wider",
    "widest",
    "playing",
    "played",
    "player",
    "players"
]

print(corpus)
```

```python
# Build vocabulary with character-level tokens

vocab = {}

for word in corpus:
    chars = tuple(list(word) + ["</w>"])
    vocab[chars] = vocab.get(chars, 0) + 1

print("Vocabulary:")
for word, freq in vocab.items():
    print(word, ":", freq)
```

```python
# Count adjacent symbol pairs

pairs = {}

for word, freq in vocab.items():
    symbols = list(word)

    for i in range(len(symbols) - 1):
        pair = (symbols[i], symbols[i + 1])

        if pair not in pairs:
            pairs[pair] = 0

        pairs[pair] += freq

print("Adjacent Pair Frequencies:\n")

for pair, count in pairs.items():
    print(pair, ":", count)
```

```python
# Find the most frequent adjacent pair

best_pair = max(pairs, key=pairs.get)

print("Most Frequent Pair:")
print(best_pair)

print("\nFrequency:")
print(pairs[best_pair])
```

```python
# Merge the most frequent pair

def merge_vocab(pair, vocab):
    new_vocab = {}

    bigram = pair

    for word, freq in vocab.items():
        new_word = []
        i = 0

        while i < len(word):
            if (
                i < len(word) - 1
                and word[i] == bigram[0]
                and word[i + 1] == bigram[1]
            ):
                new_word.append(word[i] + word[i + 1])
                i += 2
            else:
                new_word.append(word[i])
                i += 1

        new_vocab[tuple(new_word)] = freq

    return new_vocab


vocab = merge_vocab(best_pair, vocab)

print("Vocabulary After One Merge:\n")

for word, freq in vocab.items():
    print(word, ":", freq)
```

```python
# Merge the most frequent pair

def merge_vocab(pair, vocab):
    new_vocab = {}

    for word, freq in vocab.items():
        new_word = []
        i = 0

        while i < len(word):

            if (
                i < len(word) - 1
                and word[i] == pair[0]
                and word[i + 1] == pair[1]
            ):
                new_word.append(word[i] + word[i + 1])
                i += 2
            else:
                new_word.append(word[i])
                i += 1

        new_vocab[tuple(new_word)] = freq

    return new_vocab


vocab = merge_vocab(best_pair, vocab)

print("Vocabulary After Merge:\n")

for word, freq in vocab.items():
    print(word, ":", freq)
```

```python
# Function to count adjacent symbol pairs

def get_pair_frequencies(vocab):
    pairs = {}

    for word, freq in vocab.items():
        for i in range(len(word)-1):
            pair = (word[i], word[i+1])

            if pair not in pairs:
                pairs[pair] = 0

            pairs[pair] += freq

    return pairs


# Number of merges
num_merges = 10

merge_rules = []

for i in range(num_merges):

    pairs = get_pair_frequencies(vocab)

    if not pairs:
        break

    best_pair = max(pairs, key=pairs.get)

    merge_rules.append(best_pair)

    print(f"Merge {i+1}: {best_pair} Frequency = {pairs[best_pair]}")

    vocab = merge_vocab(best_pair, vocab)
```

```python
print("Learned Merge Rules:\n")

for i, rule in enumerate(merge_rules, start=1):
    print(f"{i}. {rule}")
```

```python
print("Final Vocabulary:\n")

for word, freq in vocab.items():
    print(word, ":", freq)
```

```python
# Build final vocabulary

final_vocab = set()

for word in vocab:
    for token in word:
        final_vocab.add(token)

print("Final Vocabulary:\n")

for token in sorted(final_vocab):
    print(token)

print("\nVocabulary Size:", len(final_vocab))
```

```python
def encode(word, merge_rules):

    tokens = list(word) + ["</w>"]

    for pair in merge_rules:

        i = 0
        new_tokens = []

        while i < len(tokens):

            if (
                i < len(tokens)-1
                and tokens[i] == pair[0]
                and tokens[i+1] == pair[1]
            ):
                new_tokens.append(pair[0] + pair[1])
                i += 2
            else:
                new_tokens.append(tokens[i])
                i += 1

        tokens = new_tokens

    return tokens
```

```python
def decode(tokens):

    word = ""

    for token in tokens:

        if token != "</w>":
            word += token.replace("</w>", "")

    return word
```

```python
test_word = "lowest"

encoded = encode(test_word, merge_rules)

print("Original Word :", test_word)
print("Encoded Tokens :", encoded)

decoded = decode(encoded)

print("Decoded Word :", decoded)
```

```python
test_words = [
    "lower",
    "widest",
    "players",
    "newer",
    "playing"
]

for word in test_words:

    encoded = encode(word, merge_rules)

    decoded = decode(encoded)

    print("-" * 40)
    print("Original :", word)
    print("Encoded  :", encoded)
    print("Decoded  :", decoded)
```
