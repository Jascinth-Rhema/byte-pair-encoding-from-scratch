```python
# ================================
# STEP 1 : Import Libraries
# ================================

import numpy as np
import re
from collections import Counter, defaultdict

# ================================
# STEP 2 : Create Text Corpus
# ================================

corpus = """
Machine learning is transforming the world.
Machine learning uses data to make predictions.
Artificial intelligence helps solve complex problems.
Natural language processing allows computers to understand text.
Deep learning uses neural networks.
Python is widely used for machine learning.
Byte Pair Encoding is a subword tokenization algorithm.
Large language models predict the next token.
Autoregressive models generate text one token at a time.
Data science combines statistics and programming.
Transformers use self attention mechanisms.
AI systems continue to improve every year.
"""

# ================================
# STEP 3 : Preprocess Text
# ================================

text = corpus.lower()

# Remove punctuation
text = re.sub(r'[^a-z\s]', '', text)

# Remove extra spaces
text = " ".join(text.split())

print("Processed Text:\n")
print(text)

# ================================
# STEP 4 : Split into Words
# ================================

words = text.split()

print("\n--------------------------------")
print("Total Words :", len(words))
print("--------------------------------")

print("\nFirst 20 Words:")
print(words[:20])

# ================================
# STEP 5 : Count Word Frequencies
# ================================

word_freq = Counter(words)

print("\n--------------------------------")
print("Word Frequencies")
print("--------------------------------")

for word, freq in word_freq.items():
    print(f"{word:15} : {freq}")

# ================================
# STEP 6 : Build Initial BPE Vocabulary
# ================================

vocab = {}

for word, freq in word_freq.items():

    chars = " ".join(list(word)) + " </w>"

    vocab[chars] = freq

print("\n--------------------------------")
print("Initial BPE Vocabulary")
print("--------------------------------")

for word, freq in vocab.items():
    print(f"{word} : {freq}")
```

```python
# ==========================================
# Find Adjacent Character Pair Frequencies
# ==========================================

def get_pair_frequencies(vocab):
    pair_freq = defaultdict(int)

    for word, freq in vocab.items():

        symbols = word.split()

        for i in range(len(symbols)-1):

            pair = (symbols[i], symbols[i+1])

            pair_freq[pair] += freq

    return pair_freq


pair_freq = get_pair_frequencies(vocab)

print("Adjacent Character Pair Frequencies:\n")

for pair, freq in pair_freq.items():
    print(pair, ":", freq)
```

```python
best_pair = max(pair_freq, key=pair_freq.get)

print("Most Frequent Pair :", best_pair)
print("Frequency :", pair_freq[best_pair])
```

```python
# ======================================
# Complete Byte Pair Encoding Algorithm
# ======================================

import re

def get_pair_frequencies(vocab):
    pair_freq = defaultdict(int)

    for word, freq in vocab.items():

        symbols = word.split()

        for i in range(len(symbols)-1):
            pair = (symbols[i], symbols[i+1])
            pair_freq[pair] += freq

    return pair_freq


def merge_vocab(pair, vocab):

    new_vocab = {}

    bigram = re.escape(" ".join(pair))
    replacement = "".join(pair)

    pattern = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')

    for word in vocab:
        new_word = pattern.sub(replacement, word)
        new_vocab[new_word] = vocab[word]

    return new_vocab


# Number of merges
num_merges = 10

merge_rules = []

for i in range(num_merges):

    pair_freq = get_pair_frequencies(vocab)

    if not pair_freq:
        break

    best_pair = max(pair_freq, key=pair_freq.get)

    print(f"Merge {i+1}: {best_pair} -> Frequency = {pair_freq[best_pair]}")

    merge_rules.append(best_pair)

    vocab = merge_vocab(best_pair, vocab)

print("\n==============================")
print("Learned Merge Rules")
print("==============================")

for rule in merge_rules:
    print(rule)
```
